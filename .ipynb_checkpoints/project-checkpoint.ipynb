# Проект: Обучение с учителем: качество модели

## Описание проекта

Интернет-магазин «В один клик» продаёт разные товары: для детей, для дома, мелкую бытовую технику, косметику и даже продукты. Отчёт магазина за прошлый период показал, что активность покупателей начала снижаться. Привлекать новых клиентов уже не так эффективно: о магазине и так знает большая часть целевой аудитории. Возможный выход — удерживать активность постоянных клиентов. Сделать это можно с помощью персонализированных предложений.
Нам нужно разработать решение, которое позволит персонализировать предложения постоянным клиентам, чтобы увеличить их покупательскую активность.

Нужно построить модель, которая предскажет вероятность снижения покупательской активности клиента в следующие три месяца.
В исследование нужно включить дополнительные данные финансового департамента о прибыльности клиента: какой доход каждый покупатель приносил компании за последние три месяца.
Используя данные модели и данные о прибыльности клиентов, нужно выделить сегменты покупателей и разработать для них персонализированные предложения.

## Описание данных

Таблица `market_file.csv` содержит данные о поведении покупателя на сайте, о коммуникациях с покупателем и его продуктовом поведении. Столбцы таблицы:  
`id` — номер покупателя в корпоративной базе данных.  
`Покупательская активность` — рассчитанный класс покупательской активности (целевой признак): «снизилась» или «прежний уровень».  
`Тип сервиса` — уровень сервиса, например «премиум» и «стандарт».  
`Разрешить сообщать` — информация о том, можно ли присылать покупателю дополнительные предложения о товаре. Согласие на это даёт покупатель.  
`Маркет_актив_6_мес` — среднемесячное значение маркетинговых коммуникаций компании, которое приходилось на покупателя за последние 6 месяцев. Это значение показывает, какое число рассылок, звонков, показов рекламы и прочего приходилось на клиента.  
`Маркет_актив_тек_мес` — количество маркетинговых коммуникаций в текущем месяце.  
Длительность — значение, которое показывает, сколько дней прошло с момента регистрации покупателя на сайте.  
`Акционные_покупки` — среднемесячная доля покупок по акции от общего числа покупок за последние 6 месяцев.  
`Популярная_категория` — самая популярная категория товаров у покупателя за последние 6 месяцев.  
`Средний_просмотр_категорий_за_визит` — показывает, сколько в среднем категорий покупатель просмотрел за визит в течение последнего месяца.  
`Неоплаченные_продукты_штук_квартал` — общее число неоплаченных товаров в корзине за последние 3 месяца.  
`Ошибка_сервиса` — число сбоев, которые коснулись покупателя во время посещения сайта.  
`Страниц_за_визит` — среднее количество страниц, которые просмотрел покупатель за один визит на сайт за последние 3 месяца.  
Таблица `market_money.csv` с данными о выручке, которую получает магазин с покупателя, то есть сколько покупатель всего потратил за период взаимодействия с сайтом. В таблице следующие столбцы:  
`id` — номер покупателя в корпоративной базе данных.  
`Период` — название периода, во время которого зафиксирована выручка. Например, 'текущий_месяц' или 'предыдущий_месяц'.  
`Выручка` — сумма выручки за период.  
Таблица `market_time.csv` с данными о времени (в минутах), которое покупатель провёл на сайте в течение периода. Столбцы таблицы:  
`id` — номер покупателя в корпоративной базе данных.  
`Период` — название периода, во время которого зафиксировано общее время.  
`минут` — значение времени, проведённого на сайте, в минутах.  
Таблица `money.csv` с данными о среднемесячной прибыли покупателя за последние 3 месяца: какую прибыль получает магазин от продаж каждому покупателю. В таблице следующие столбцы:  
`id` — номер покупателя в корпоративной базе данных.  
`Прибыль` — значение прибыли.

Загрузим данные из файлов в датафреймы и сохраним их во вложенные списки списка `dfs` вместе с соответствующими описаниями таблиц.

```python
dfs = ([['данные о поведении на сайте, о коммуникациях и продуктовом поведении покупателя'], 
        ['данные о выручке, которую получает магазин с покупателя'], 
        ['данные о времени на сайте'],
        ['данные о среднемесячной прибыли покупателя за последние 3 месяца']]
)

alt_urls = (
    'https://code.s3.yandex.net/datasets/market_file.csv',
    'https://code.s3.yandex.net/datasets/market_money.csv',
    'https://code.s3.yandex.net/datasets/market_time.csv'
    'https://code.s3.yandex.net/datasets/money.csv'
)

urls = (
    'market_file.csv',
    'market_money.csv',
    'market_time.csv',
    'money.csv'
)

# загрузим данные в таблицы и сохраним таблицы во вложенные списки списка dfs после их описаний 
for i, url in enumerate(urls):
    if i == 3:
        sep = ';'
        decimal = ','
    else:
        sep = ','
        decimal = '.'
    try:
        dfs[i].append(pd.read_csv(url, sep=sep, decimal=decimal))
    except:
        dfs[i].append(pd.read_csv(alt_urls[i], sep=sep, decimal=decimal))

# трансформируем вложенные списки списка dfs в кортежи        
for i in range(len(dfs)):
    dfs[i] = tuple(dfs[i])
```

Для знакомства с данными используем вспомогательные функции.

```python
# напишем функции для правописания слов "столбец" и "строка"
def row_word(row_num):
    if str(row_num)[-1] == '1':
        return 'строка'
    elif str(row_num)[-1] in '234' and row_num not in [11, 12, 13, 14]:
        return 'строки'
    else:
        return 'строк'

def col_word(col_num):
    if str(col_num)[-1] == '1':
        return 'столбец'
    elif str(col_num)[-1] in '234' and col_num not in [11, 12, 13, 14]:
        return 'столбца'
    else:
        return 'столбцов'

# напишем функцию для знакомства с датафреймом
def df_inf(df, description):
    print(f'\nРассмотрим датафрейм, содержащий {description}.\n')
    df.info()
    display(df.head())

    # подсчет строковых столбцов
    obj_col_num = 0
    for i in df.columns:
        if is_string_dtype(df[i]):
            obj_col_num += 1

    # подсчет строк и столбцов
    row_num = df.shape[0]
    col_num = df.shape[1]
    
    print(f'\nВ датафрейме {row_num} {row_word(row_num)} и {col_num} {col_word(col_num)}. Из них {obj_col_num}'
          f' {col_word(obj_col_num)} со строковыми значениями и {col_num - obj_col_num} с числовыми.')
    
    # подсчет столбцов с пропусками
    nan_col_num = 0
    for i in df.columns:
        if df[i].isna().sum() > 0:
            nan_col_num += 1
   
    if nan_col_num == 0:
        print(f'Пропущенных значений в датафрейме нет.')
    else:
        print(f'В {nan_col_num} {col_word(nan_col_num)} есть пропущенные значения.')
    print('-------------------------------------------------')
```

Взглянем на данные из четырех датафреймов.

```python
for x in dfs:
    df_inf(x[1], x[0])
```

На первый взгляд данные соответствуют описаниям, есть неточности в написании некоторых категорий, что может сигнализировать о наличии неявных дубликатов в значениях этих категорий.

##  Предобработка данных

Как мы уже заметили, в данных нет пропущенных значений. Проверим таблицы на наличие дубликатов строк.

```python
for df in dfs:
    if df[1].duplicated().sum() > 0:
        print(f'В таблице, содержащей {df[0]}, найдено дубликатов: {df[1].duplicated().sum()}.')
    else:
        print(f'В таблице, содержащей {df[0]}, дубликатов нет.')
```

Выясним, есть ли проблемы со значениями категориальных данных в наших таблицах.

```python
# выведем уникальные значения категориальных признаков
for df in dfs:
    print(f'\nВ таблице, содержащей {df[0]}:\n')
    cat_cols = list(df[1].select_dtypes(include='object'))
    if cat_cols == []:
        print('Категориальных признаков нет.')
    for cat in cat_cols:
        print(f'Уникальные значения признака {cat}: {list(df[1][cat].unique())}')
```

Мы видим, что необходимо привести к единому написанию значения признаков `Тип сервиса` и `Период`.

```python
# приведем к единому написанию значение "стандарт" признака "Тип сервиса"
dfs[0][1]['Тип сервиса'] = dfs[0][1]['Тип сервиса'].apply(lambda x: 'стандарт' if x=='стандартт' else x)
dfs[0][1]['Тип сервиса'].unique()
```

```python
# приведем к правильному написанию значение "предыдущий_месяц" признака "Период"
dfs[2][1]['Период'] = dfs[2][1]['Период'].apply(lambda x: 'предыдущий_месяц' if x=='предыдцщий_месяц' else x)
dfs[2][1]['Период'].unique()
```

**Таким образом, мы ознакомились с данными, и выяснили что в них нет пропусков и явных дубликатов. Далее, мы избавились от неявных дубликатов в значениях признаков `Тип сервиса` и `Период` путем приведения этих значений к единому написанию.**

## Исследовательский анализ данных

Для исследовательского анализа данных используем вспомогательную функцию:

```python
def describe_feature(df, column, bins=20, as_cat=False):
    print(f"\n          === Признак '{column}' ===")
    
# для строковых данных выведем частотную таблицу и круговую диаграмму или столбчатый график
# при as_cat=True количественный признак описывается аналогично категориальному
    if is_string_dtype(df[column]) or as_cat:
        
        df_value_counts = df[column].value_counts()
        display(df_value_counts)
        
        fig, ax = plt.subplots(figsize=(9, 6));
        if len(df_value_counts) < 4:
            df_value_counts.plot(kind='pie', autopct="%.2f%%");
            plt.title(f'Круговая диаграмма признака {column}');
            plt.ylabel('');
            plt.show()
        else:
            df_value_counts.plot(kind='bar');
            plt.title(f'Столбчатый график признака {column}');
            plt.show()

# для категориальных данных опишем характеристики частотной таблицы
        df_value_counts_sorted = df_value_counts.sort_values(ascending=False)
        description = (f"Значениями столбца '{column}' являются: " 
                       + str(list(df_value_counts.index))[1:-1] + ". ")
        for i, x in enumerate(df_value_counts_sorted):
            if i == 0:
                description += (f"Чаще всего ({x} {row_word(x)} или {x/len(df):.2%} наблюдений) признак '{column}'"
                                f" принимает значение '{df_value_counts_sorted.index[i]}'")
            elif i == len(df_value_counts_sorted) - 1:
                description += (f", в наименьшем количестве наблюдений ({x} {row_word(x)}  или {x/len(df):.2%}"
                                f" наблюдений) этот признак принимает значение '{df_value_counts_sorted.index[i]}'.")
        print(description)            

# для числовых данных выведем стататистики, гистограмму и ящик с усами
    else:
        stats = df[column].describe()
        display(stats)
    
        fig, ax = plt.subplots(figsize=(9, 6));
        plt.hist(df[column], bins=bins);
        plt.axvline(x=df[column].median(), color='red', label='Медиана');
        plt.axvline(x=df[column].mean(), color='red', label='Среднее', linestyle='--');
        plt.legend()
        plt.title(f'Гистограмма признака {column}')
        plt.show()
    
        fig, ax = plt.subplots(figsize=(9, 6));
        plt.boxplot(df[column], vert=False);
        plt.title(f'График ящик с усами пизнака {column}')
        plt.show()
        
# для количественных данных опишем статистические характеристики распределения
        print(f"Количественный признак '{column}' принимает значения в диапазоне от {stats.loc['min']}"
              f" до {stats.loc['max']:.2f}. Среднее значение признака '{column}' - {stats.loc['mean']:.2f},"
              f" медианное - {stats.loc['50%']:.2f}. Величина стандартного отклонения: {stats.loc['std']:.2f}.")
```

По ТЗ для дальнейшего анализа нам необходимо отобрать клиентов с покупательской активностью не менее трёх месяцев, то есть таких, которые что-либо покупали в этот период, то есть клиентов у которых были покупки в предыдущий и препредыдущий месяца.

```python
mask = (((market_money['Период'] == 'предыдущий_месяц') | (market_money['Период'] == 'препредыдущий_месяц'))
         & (market_money['Выручка'] == 0))

clients_to_drop = market_money[mask]
clients_to_drop
```

Как мы видим, есть три клиента у которых не было покупок в предыдущий и препредыдущий месяца. Удалим данные о них из дальнейшего рассмотрения.

```python
market = market.drop(market[market['id'].isin(clients_to_drop['id'])].index)
```

Можно заметить, что данные о выручке и времени на сайте находятся в одном столбце для всех периодов, создадим отдельный столбец для каждого периода.

```python
# создадим отдельные столбцы по месяцам
market_money = pd.pivot_table(market_money, index='id', columns='Период', values='Выручка')
market_time = pd.pivot_table(market_time, index='id', columns='Период', values='минут')

# переименуем столбцы
market_money.columns = ['предыдущий_месяц_выручка', 'препредыдущий_месяц_выручка', 'текущий_месяц_выручка']
market_time.columns = ['предыдущий_месяц_минут', 'текущий_месяц_минут']
```

## Пайплайн

```python
# установим id в качестве индекса
data = data.set_index('id')

# формируем выборки со стратификацией
X_train, X_test, y_train, y_test = train_test_split(
    data.drop(['Покупательская активность'], axis=1),
    data['Покупательская активность'],
    test_size = TEST_SIZE, 
    random_state = RANDOM_STATE,
    stratify = data['Покупательская активность'])
```

```python
print(f'Размерность исходной выборки {data.shape}'
      f'\nРазмерность тренировочной выборки {X_train.shape}'
      f'\nРазмерность тестовой выборки {X_test.shape}')
```

Разбиение осуществлено верно.

Подберем кодировщики для категориальных данных.

```python
for col in cat_col_names:
    print(col)
    display(data[col].unique())
```

Значения признака `Тип сервиса` предполагают ранжирование (в некотором смысле 'премиум' "лучше" 'стандарта'), поэтому будет разумно закодировать его с помощью `OrdinalEncoder`. Значения категориальных признаков `Разрешить сообщать` и `Популярная_категория` не подразумевают ранжирования, следовательно их лучше закодировать с использованием `OneHotEncoder`.

```python
# создаём списки с названиями признаков для OneHotEncoder и OrdinalEncoder
ohe_columns = ['Разрешить сообщать', 'Популярная_категория']
ord_columns = ['Тип сервиса']
```

Закодируем значения целевого признака с помощью `LabelEncoder`.

```python
le = LabelEncoder()
y_train = le.fit_transform(y_train)
y_test = le.transform(y_test)

print(f"Значение '{le.inverse_transform([0, 1])[0]}' закодировано как '0', '{le.inverse_transform([0, 1])[1]}' - как '1'.")
```

Далее, создадим пайплайн, заполняющий потенциальные пропуски, кодирующий категориальные данные и масштабирующий количественные данные. Будем перебирать пять моделей (KNeighborsClassifier(), DecisionTreeClassifier(), LogisticRegression(), SVC(), GaussianNB()) с подбором гиперпараметров и скейлеров (MinMaxScaler(), StandardScaler() или данные без масштабирования) с помощью `GridSearchCV`.

```python
# создаём пайплайн для подготовки признаков из списка ohe_columns: заполнение пропусков и OHE-кодирование
# SimpleImputer + OHE
ohe_pipe = Pipeline(
    [('simpleImputer_ohe', SimpleImputer(missing_values=np.nan, strategy='most_frequent')),
     ('ohe', OneHotEncoder(drop='first', handle_unknown='ignore', sparse_output=False))
    ]
    )

# создаём пайплайн для подготовки признаков из списка ord_columns: заполнение пропусков и Ordinal-кодирование
# SimpleImputer + OE
ord_pipe = Pipeline(
    [('simpleImputer_before_ord', SimpleImputer(missing_values=np.nan, strategy='most_frequent')),
     ('ord',  OrdinalEncoder(
                categories=[
                    ['стандарт', 'премиум']
                ], 
                handle_unknown='use_encoded_value', unknown_value=np.nan
            )
        ),
     ('simpleImputer_after_ord', SimpleImputer(missing_values=np.nan, strategy='most_frequent'))
    ]
)

# создаём общий пайплайн для подготовки данных
data_preprocessor = ColumnTransformer(
    [('ohe', ohe_pipe, ohe_columns),
     ('ord', ord_pipe, ord_columns),
     ('num', MinMaxScaler(), num_col_names)
    ], 
    remainder='passthrough'
)

# создаём итоговый пайплайн: подготовка данных и модель
pipe_final = Pipeline([
    ('preprocessor', data_preprocessor),
    ('models', DecisionTreeClassifier(random_state=RANDOM_STATE))
])

param_grid = [
# словарь для модели DecisionTreeClassifier()
    {
        'models': [DecisionTreeClassifier(random_state=RANDOM_STATE)],
        'models__max_depth': range(2, 5),
        'models__max_features': range(2, 5),
        'preprocessor__num': [StandardScaler(), MinMaxScaler(), 'passthrough']  
    },
    
# словарь для модели KNeighborsClassifier() 
    {
        'models': [KNeighborsClassifier()],
        'models__n_neighbors': range(2, 5),
        'preprocessor__num': [StandardScaler(), MinMaxScaler(), 'passthrough']   
    },

# словарь для модели LogisticRegression()
    {
        'models': [LogisticRegression(
            random_state=RANDOM_STATE, 
            solver='liblinear', 
            penalty='l1'
        )],
        'models__C': range(1, 10),
        'preprocessor__num': [StandardScaler(), MinMaxScaler(), 'passthrough']  
    },

# словарь для модели SVC()
    {
        'models': [SVC(
            random_state=RANDOM_STATE,
            kernel='poly'
        )],
        'models__C': range(1, 5),
        'models__degree': range(2, 5),
        'preprocessor__num': [StandardScaler(), MinMaxScaler(), 'passthrough']  
    },

# словарь для модели GaussianNB()
    {
        'models': [GaussianNB()],
        'preprocessor__num': [StandardScaler(), MinMaxScaler(), 'passthrough']
    }
]

grid_search = GridSearchCV(
    pipe_final, 
    param_grid, 
    cv=5,
    scoring='roc_auc',
    n_jobs=-1
)
```

Для выбора лучшей модели и гиперпараметров будем использовать метрику `roc_auc`, поскольку она подходит для оценки всех перебираемых моделей (при всех значениях порогов вероятности присваивания классов, где это имеет смысл).

```python
grid_search.fit(X_train, y_train)

print('Лучшая модель и её параметры:\n\n', grid_search.best_estimator_)
display(grid_search.best_estimator_)
print ('Метрика лучшей модели по результатам кроссвалидации на тренировочной выборке:', grid_search.best_score_)

# проверим работу модели на тестовой выборке
# рассчитаем прогноз на тестовых данных
y_test_pred = grid_search.predict(X_test)
print(f'Метрика ROC-AUC на тестовой выборке:'
      f' {roc_auc_score(y_test, grid_search.predict_proba(X_test)[:, 1])}')
```

Модель показывает хорошие результаты на тестовой выборке, хоть и заметен небольшой эффект переобучения.

**Таким образом, с помощью инструментов кросс-валидации нами была выбрана модель, показывающая лучшие значения метрики `roc_auc`= 0.9082 на тренировочной выборке. Это модель логистической регрессии с полиномиальным ядром, l1-регуляризаций с параметром C=1 и масштабированием количественных признаков с помощью `MinMaxScaler()`.**

## Анализ важности признаков

Проведем анализ важности признаков полученной модели с помощью методов библиотеки `shap`.

```python
# трансформируем тренировочные данные отобранным методом
X_train_transformed = pd.DataFrame(grid_search.best_estimator_.named_steps['preprocessor'].fit_transform(X_train))
X_train_transformed.columns = grid_search.best_estimator_.named_steps['preprocessor'].get_feature_names_out()

# создадим объект LinearExplainer, с лучшей моделью и трансформированными данными в качестве параметров
explainer = shap.LinearExplainer(grid_search.best_estimator_.named_steps['models'], X_train_transformed)

# оценим shap-значения и построим график общей значимости признаков
shap_values = explainer(X_train_transformed)
shap.plots.bar(shap_values, max_display=17)
```
